```python
#Import needed libraries
import pandas as pd
import numpy as np
from deep_translator import GoogleTranslator
```

```python
#Load the dataset here
df = pd.read_csv('RawData.csv')
```

```python
#Display first 5 values
df.head()
```

```python
# Convert 'TIME' column to timedelta. This is for filtering under 15 minute surveys
try:
    df['TIME'] = pd.to_timedelta(df['TIME'], errors='coerce')
    print(df['TIME'].dtype)
    print(df.head())
except Exception as e:
    print(f"An error occurred during datetime conversion: {e}")
```

```python
# Filter STA column with 'COMP'
filtered_df = df[df['STA'] == 'COMP']
print(filtered_df.head)
```

```python
# Count 'COMP' values in STA column in filtered_df

comp_count = len(filtered_df[filtered_df['STA'] == 'COMP'])
print(f"Number of 'COMP' values in the 'STA' column of the filtered DataFrame: {comp_count}")
```

```python
# Check for duplicates in the second column ('mid') of filtered_df
duplicates = filtered_df.duplicated(subset=filtered_df.columns[1], keep=False)

if duplicates.any():
    # Delete rows with duplicates, keeping the first occurrence
    filtered_df = filtered_df[~duplicates]
    print("Duplicates removed from the DataFrame.")
else:
    print("No duplicates found in the second column.")

filtered_df.head()
```

```python
#Remove rows with time under 15 mins
filtered_df[filtered_df['TIME'] >= pd.Timedelta(minutes=15)]

removed_count = filtered_df.shape[0] - filtered_df.shape[0]
print(f"Rows removed (under 15 mins): {removed_count}")
```

```python
#concatenate 'UserAgent' and 'IPAddress' in the filtered_df dataframe, assign new column name 'BrowserInfo'
if 'UserAgent' in filtered_df.columns and 'IPAddress' in filtered_df.columns:
    filtered_df['BrowserInfo'] = filtered_df['UserAgent'] + ' ' + filtered_df['IPAddress']
    print(filtered_df.head())
else:
    print("Either 'UserAgent' or 'IPAddress' column not found in the DataFrame.")
```

```python
#Determine if there are duplicates within the 'BrowserInfo' column of filtered_df dataframe

duplicates_browser = filtered_df.duplicated(subset=['BrowserInfo'], keep=False)

if duplicates_browser.any():
    print("Duplicate BrowserInfo entries found.")
    # You can further process the duplicates here, e.g., display them:
    print(filtered_df[duplicates_browser])
else:
    print("No duplicate BrowserInfo entries found.")
```

```python
# Apply translation. List the columns you want to translate
columns_to_translate = ['Q2t1','Q9At2','Q12t1','Q19t1','Q20t1','Q21','Q22t1','Q23','Q24t1'] # please check column names before proceeding

# Define a reusable translator function
translator = GoogleTranslator(source='auto', target='en')
def translate_text(text):
    try:
        return translator.translate(str(text))
    except Exception:
        return text  # return original if translation fails

# Apply translation to selected columns
for col in columns_to_translate:
    filtered_df[col] = filtered_df[col].apply(translate_text)
```

```python
#Export the filtered_df dataframe, convert to a downloadable csv file and rename as 'NewData.csv'

from google.colab import files
filtered_df.to_csv('NewData.csv', encoding = 'utf-8-sig')
files.download('NewData.csv')
```
